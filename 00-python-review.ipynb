### You can access _individual characters in a string_ using `indexes`

```python
greeting = 'happy birthday'
```

To use indexing, you must use the square brackets like this:

```python
# Below, 'greeting' is the name of the variable containing our string.
# We follow that with the square brackets. Between the square brackets, 
# we write the position of the character we want.

greeting[0]
```

```python
greeting[1]
```

```python
greeting[2]
```

```python
greeting[-1]
```

```python
greeting[-2]
```

### You can access _parts of a string_ using `indexes`

- this is also called `slicing`

```python
print(greeting)
```

The code below returns all characters starting at index 0, all the way up to, but EXCLUDING index 3.

```python
# Below, 'greeting' is the name of the variable containing our string.
# We follow that with the square brackets. Between the square brackets, 
# we write the position of the character we want start from (0), followed
# by ':', followed by the position of the character we want to end before (3).

greeting[0:3]
```

```python
greeting[2:4]
```

If you leave out the first number, it will return all the characters starting from the beginning.

```python
# This returns all characters starting at index 0,
# going all the way up to, but EXCLUDING the character at index 5.
# In essence, this returns the part of the string composed of the
# first 5 characters.

greeting[:5]
```

If you leave out the second number, it will return all the characters until the end of the string.

```python
# This returns all characters starting at index 1, 
# going all the way until the end of the string.
# In essence, this returns the entire string, except for the first
# character.

greeting[1:]
```

### NOTES:

- **Remember, for indexing, counting starts at 0**
- In programming, in general, we always start counting at 0

# Tuples Data Type

* **ordered sequences** that contains items (objects) of various data types

* often wrapped between `( )`

* objects inside are comma separated

* can contain other lists, dictionaries, tuples etc.

* can use indexing and slicing to extract items from lists

**NOTE: Basically, they're very similar to lists, except:**
  - you cannot add items to a tuple
  - you cannot change an item in a tuple
  - you cannot delete an item from a tuple

```python
my_tuple = (1,2,3,4,5)
```

```python
my_tuple[0]
```

```python
empty_tuple = ()
```

**NOTE: Single item tuples still require the comma:**

```python
one_item_tuple = ("a",) # this is a tuple
```

```python
not_a_tuple = ("a") # this is NOT a tuple; it's equivalent to the string "a"
```

```python
t = 1, 2,
type(t)
```

**Question:** Why do we have both lists and tuples?

**Answer:** Tuples are less flexible than lists, however they are faster and occupy less memory. So, if you have ordered sequences of objects that you don't need to change, use a `tuple` - it will be faster and use up less memory. If you have ordered sequences of objects that you DO need to change, use a `list`.

## Classes

* in `Python`, you can use classes to create blueprints for new objects

* classes essentially describe what kind of attributes and methods a type (or class) of objects has available

```python
class Car():
    
    # We use __init__ to describe how a new car object
    # is created from given inputs. The first parameter
    # is `self` and points to the object being built.
    
    def __init__(self, make, model, license_plate):
        self.make = make
        self.model = model
        self.license_plate = license_plate
    
    
    # This is an instance method. Every object created
    # has a `drive` method. Instance methods can take
    # any number of parameters. In this example, we take
    # a parameter called `miles`, which represents the 
    # number of miles traveled.
    
    def drive(self, miles):
        print(f'{self.make} drove {miles} miles.')
        
        
    def stop(self):
        print(f'{self.make} is now stopped.')
```

```python
# This is how we create a new object, using the 
# `Car` class (or blueprint)

my_car = Car('Saab', '9-3', '5555')
```

```python
# This object now has a make, model and a license plate

my_car.make
```

```python
my_car.model
```

```python
my_car.license_plate
```

```python
# This object also has the `drive` and `stop` methods available.

# Notice that when you call an instance method, you DON'T
# have to pass a value for the `self` parameter - Python does
# that for you. You only need to specify the other parameters.

# For example, the method `drive` takes one parameter, `miles`, 
# which represents the number of miles traveled.

my_car.drive(150)
```

```python
my_car.stop()
```

# Conditionals

## `if... elif... else`

- sometimes we need to execute some code **if and only if** something is `True` (or `False`)
  - e.g. write a file **if and only if** it doesn't exist

- Python has a built-in code structure we can use to create such code

```python
# The code below will print 'Turn heat on'. The way it works is as follows:
# the variable 'temperature' will be compared with 71. Because temperature = 59,
# the expression 'temperature < 71' will be True, and so the code INDENTED underneath
# the 'if' statement will run, printing 'Turn heat on'.

temperature = 59

if temperature < 71:
    print('Turn heat on')
```

We use an `elif` branch to execute code when the condition in the `if` statement is `False`.

```python
# The code below will print 'Turn heat off'. The way it works is as follows:
# the variable 'temperature' will be compared with 71. Because temperature = 75,
# the expression 'temperature < 71' will be False, and so next we'll look at the 
# condition on the 'elif' branch. In this case, 'temperature > 71' is True, so
# the code INDENTED underneat the 'elif' branch will execute, printing
# 'Turn heat off'.

temperature = 75

if temperature < 71:
    print('Turn heat on')
elif temperature > 71:
    print('Turn heat off')
```

**NOTES:**
- you can have as many `elif` branches as you need associated with ONE `if` branch
- you can have a default `else` branch that will execute if none of the other branches executed
- you can only have ONE `else` branch

```python
# In the code below, try changing 'current_day' to 'Monday', 'Wednesday' and 'Friday'.
# Observe the output.

current_day = 'Monday'

if current_day == 'Monday':
    print('meh')
elif current_day == 'Wednesday':
    print('hump day')
elif current_day == 'Thursday' or current_day == 'Friday':
    print('almost weekend')
else:
    print('yay')
```

- Conditional statements can be **nested**.

```python
weather_is_nice = False
current_day = "Tuesday"

if weather_is_nice:
    if (current_day == "Saturday" or current_day == "Sunday"):
        print("go out")
    else:
        print("go to work")
else:
    print("stay home")
```

The really powerful thing about the `if / elif / else` branches is that they can be nested! In the example above, we have a top level condition that checks if the weather is nice. If `weather_is_nice` is `False`, we end up printing 'stay home', regardless of the value of `current_day`. But if `weather_is_nice` is `True`, then we print different messages, depending on whether `current_day` is a weekend or a weekday.

# Errors and Exceptions

## You can handle errors and exceptions with the  `try` ... `except` ... `finally` statements.

* wrap the code that might throw an exception between `try`...`except`
* be specific about which exception you want to handle (e.g. `ValueError`, `KeyError`, `IndexError`, etc.)
* the code in the optional `finally` block executes at the end, **whether an exception occurs or not**

```python
# In this code example, we have a list of words and a list of integers. 
# We build a sentence by taking each of the integers and finding the word
# at that index in the list of words. If we get an `IndexError` exception, 
# we simply skip that index. In the end, we print the sentence we obtain.

words = ['the sun', 'by the lake', 'is running', 'holding', 'trees', 'the moon']
indexes = [4, 3, 1, 2, 0, 7]

try:
    sentence = ''
    for i in indexes:
        sentence += f'{words[i]} '
except IndexError:
    pass
finally:
    print(sentence)

print('Finished.')
```

```python
# Notice what happens when we don't handle the exception

words = ['the sun', 'by the lake', 'is running', 'holding', 'trees', 'the moon']
indexes = [4, 3, 1, 2, 0, 7]

try:
    sentence = ''
    for i in indexes:
        sentence += f'{words[i]} '
finally:
    print(sentence)

print('Finished.')
```

## You can raise exceptions

You can use `raise` to raise a new exception. You can then, in other parts of the code, handle this exception.

```python
# In the code below, we raise an exception 
# if a given password contains no numbers.

def has_digits(text):
    for i in range(9):
        if str(i) in text:
            return True
    
    return False

def set_password(password):
    if not has_digits(password):
        raise Exception('Password is too weak.')
    else:
        print('Password set successfully.')
```

```python
set_password('pa55w0rd4')
```

```python
set_password('password')
```

# Pickling

## **Writing / reading native Python objects (pickling)**

- Some Python objects take a long time to obtain (e.g. a neural network model)
<br><br>
- Some Python objects need to be backed up
<br><br>
- We can save them to a file and restore them when we need them

## Pickle
- A method of "serialization" of objects (representing them in a way that is generally understood across platforms) 
<br><br>
- It is accessed through the `import pickle` command

```python
a = 1
```

```python
import pickle
```

```python
x = pickle.dumps(a)
```

```python
del a
```

```python
print(x)
```

```python
y = pickle.loads(x)
```

```python
print(y)
```

```python
print(a)
```

**In the code above, we:**
 - create an integer-valued variable
     - export it to a byte representation using `pickle.dumps()`
 - delete the initial variable
 - print the pickled bytes representation we created
 - load the bytes representation into a new variable using pickle.loads()
 - confirm that the __value__ of the saved variable was correctly read
 - confirm that the initial variable is NOT retrieved via this save/reload

## You can access arguments that were passed to a Python script using the built-in `sys.argv` list.

* Let's say we have a script called `my_script.py`

* Let's say we run the script from the command line using the following command:  
  `python my_script.py -n Ciprian`

* Inside your script, `sys.argv` will be equal to `['my_script.py', '-n', 'Ciprian']`
  * to get the name of the script, you can use `sys.argv[0]`
  * to get the first parameter that was passed to the script (`-n`), you can use `sys.argv[1]`
  * to get the second parameter that was passed to the script (`Ciprian`), you can use `sys.argv[2]`

* To get the total number of parameters passed to the script, you can use `len(sys.argv) - 1`
  * Why the `- 1` here? **Because the first item in the `sys.argv` list is always the name of the script file itself.**

**NOTE:** For advanced users, the `argparse` module is a far more flexible and powerful alternative to using `sys.argv`. I strongly encourage you to check it out: https://docs.python.org/3/howto/argparse.html
